**Classification**

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=40, metric='minkowski',p=2)
x= array_x
y= array_y
knn.fit(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(array_x, array_y, test_size=0.3,random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=40)
knn.fit(x_train , y_train)
y_predict = knn.predict(x_test)
y_predict
```

```python
knn.score(x_test, y_test)
```

```python
neighbors = np.arange(1,60)

train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

for i,k in enumerate(neighbors):
  knn_model = KNeighborsClassifier(n_neighbors=k)
  knn_model.fit(x_train , y_train)
  train_accuracy[i] = knn_model.score(x_train, y_train)
  test_accuracy[i] = knn_model.score(x_test, y_test)

plt.plot(neighbors,test_accuracy, label = 'Testing Accuracy')
plt.plot(neighbors, train_accuracy, label = 'Training Accuracy')
plt.legend()
plt.xlabel('number of Neighbors')
plt.ylabel('Accuracy')
plt.show()
```

**Clustering**

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
features = data2[['Metrics.Used Price', 'Metrics.Sales']]
```

```python
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(features_scaled)
```

```python
cluster_labels = kmeans.labels_
```

```python
centroids = kmeans.cluster_centers_
```

```python
data2['cluster'] = cluster_labels
```

```python
plt.scatter(features_scaled[:, 0], features_scaled[:, 1], c=cluster_labels)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=100, marker='*', linewidth=10, label='Centroids')
plt.xlabel("Feature 1 (scaled)")
plt.ylabel("Feature 2 (scaled)")
plt.title("K-Means Clustering with Centroids")
plt.legend()
plt.show()
```

```python
centroids
```

```python
kmeans.inertia_
```

```python
inertia_list = []
for k in np.arange(1,10):
  kmeans = KMeans(n_clusters = k)
  kmeans.fit(features)
  inertia_list.append(kmeans.inertia_)
inertia_list
```

```python
plt.plot(np.arange(1,10),inertia_list,'o-',c='red')
plt.xlabel('number of clusters')
plt.ylabel('Inertia')
plt.show()
```
